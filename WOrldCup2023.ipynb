# Which Country Have Highest Boundaries

```python
Boundaries = Batting.groupby('Team_Innings')[['4s','6s']].sum().reset_index()
Boundaries = Boundaries.sort_values(by = '4s', ascending = False)
```

```python
Boundaries
```

```python
plt.figure(figsize = (13,8))

ax = sns.barplot(y= '4s',x= 'Team_Innings', data = Boundaries)
ax.bar_label(ax.containers[0], fontsize = 10)
plt.title('Total 4s Boundaries', size = 20, color = 'red')
plt.xlabel('Team', size = 17, color = 'Blue')
plt.ylabel(' 4s Boundaries', size = 17, color = 'Blue')
```

```python
plt.figure(figsize = (13,8))

ax = sns.barplot(y= '6s',x= 'Team_Innings', data = Boundaries)
ax.bar_label(ax.containers[0], fontsize = 10)
plt.title('Total 6s Boundaries', size = 20, color = 'red')
plt.xlabel('Team', size = 17, color = 'Blue')
plt.ylabel(' 6s Boundaries', size = 17, color = 'Blue')
```
